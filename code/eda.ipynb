```python
from sklearn.cluster import KMeans
from tqdm import tqdm

# 위도와 경도 데이터 추출
X = train_data[['latitude', 'longitude']]

# 클러스터 수 범위 설정
cluster_range = range(1, 100)
sse = []

# 각 클러스터 수에 대해 K-means 실행 및 SSE 계산
for k in tqdm(cluster_range, desc="Calculating SSE"):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

# SSE 결과 출력
print(sse)

# 그래프 스타일 설정
plt.figure(figsize=(8, 5))

# SSE 그래프 그리기
plt.plot(cluster_range, sse, marker='o')
plt.title('Elbow Test')
plt.xlabel('Clusters')
plt.ylabel('SSE')

# 그래프 표시
plt.show()
```

```python
# 위도와 경도 데이터 추출
X = train_data[['latitude', 'longitude']]

# K-means 모델 생성 및 학습 (클러스터 수: 10)
kmeans = KMeans(n_clusters=10, random_state=42)
train_data['cluster'] = kmeans.fit_predict(X)

# 플롯 크기 설정
plt.figure(figsize=(10, 8))

# 클러스터링 결과 산점도
sns.scatterplot(x='longitude', y='latitude', hue='cluster', 
                palette='viridis', data=train_data, legend='full', alpha=0.6)

# 그래프 제목 및 레이블 설정
plt.title('K-means Clustering')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# 범례 추가 및 그래프 표시
plt.legend(title='Cluster')
plt.show()
```

```python
# 테스트 데이터의 위도와 경도 추출
X_test = test_data[['latitude', 'longitude']]

# 테스트 데이터에 클러스터 할당
test_data['cluster'] = kmeans.predict(X_test)

# 플롯 크기 설정
plt.figure(figsize=(14, 6))

# 훈련 데이터 산점도
plt.subplot(1, 2, 1)
sns.scatterplot(x='longitude', y='latitude', hue='cluster', 
                palette='viridis', data=train_data, legend='full', alpha=0.6)
plt.title('Train Data Clustering')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# 테스트 데이터 산점도
plt.subplot(1, 2, 2)
sns.scatterplot(x='longitude', y='latitude', hue='cluster', 
                palette='viridis', data=test_data, legend='full', alpha=0.6)
plt.title('Test Data Clustering')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
combined_data = pd.concat([train_data, test_data], keys=['train', 'test']).reset_index(level=0).rename(columns={'level_0': 'dataset'})

combined_data['cluster'] = kmeans.fit_predict(combined_data[['longitude', 'latitude']])

train_clustered = combined_data[combined_data['dataset'] == 'train']
test_clustered = combined_data[combined_data['dataset'] == 'test']

# 플롯 크기 설정
plt.figure(figsize=(14, 6))

# 훈련 데이터 산점도
plt.subplot(1, 2, 1)
sns.scatterplot(x='longitude', y='latitude', hue='cluster', 
                palette='viridis', data=train_clustered, legend='full', alpha=0.6)
plt.title('Train Data Clustering')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# 테스트 데이터 산점도
plt.subplot(1, 2, 2)
sns.scatterplot(x='longitude', y='latitude', hue='cluster', 
                palette='viridis', data=test_clustered, legend='full', alpha=0.6)
plt.title('Test Data Clustering')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
def categorize_area(x):
    range_start = (x // 50) * 50
    range_end = range_start + 49
    return f"{range_start} - {range_end}"

area_dummies = pd.get_dummies(train_data['area_m2'].apply(categorize_area), prefix='area',drop_first=True, dtype=int)

train_data = pd.concat([train_data, area_dummies], axis=1)

train_data.head()
```

```python
train_data['area_m2_category'] = train_data['area_m2'].apply(categorize_area)

category_means = train_data.groupby('area_m2_category')['deposit'].mean().reset_index()

plt.figure(figsize=(20, 6))
sns.barplot(x='area_m2_category', y='deposit', data=category_means)
plt.xlabel('Area Category')
plt.ylabel('Average Deposit')
plt.tight_layout()
plt.show()
```
